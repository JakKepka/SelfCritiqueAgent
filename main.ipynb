```python
from mlcroissant import Dataset

ds = Dataset(jsonld="https://huggingface.co/api/datasets/casehold/casehold/croissant")
records = ds.records("all")
```

```python
import pandas as pd

# Ensure 'records' is materialized as a list (works if it's an iterator or generator)
try:
    records_list = list(records)
except Exception:
    records_list = records

# Convert list/dict records to a DataFrame
df = pd.DataFrame(records_list)

# Quick sanity check: show type and first rows
print('records ->', type(records))
df.head()
```
